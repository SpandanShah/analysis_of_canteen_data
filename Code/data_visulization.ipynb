```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
```

```python
# read data from
weekdays = pd.read_csv('../Data/Final_Data/weekdays_Aug.csv')
weekends = pd.read_csv('../Data/Final_Data/weekend_Aug.csv')
```

```python
# count all elemet of count
weekdays_sum=weekdays['count'].sum()
```

```python
# count all elemet of count
weekend_sum=weekends['Count'].sum()
```

```python
# make a bar chart of weekdays_sum and weekend_sum
plt.bar(['Weekdays','Weekends'],[weekdays_sum,weekend_sum])
plt.title('Weekdays vs Weekends')
plt.xlabel('Days')
plt.ylabel('Count')
plt.show()
```
