```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Data/Group_data/weekdays_weekend_aug.csv')
df1 = pd.read_csv('../Data/Group_data/weekdays_weekend_sep.csv')
df2 = pd.read_csv('../Data/Group_data/weekdays_weekend_oct.csv')
df3 = pd.read_csv('../Data/Group_data/weekdays_weekend_nov.csv')
df4 = pd.read_csv('../Data/Group_data/weekdays_weekend_dec.csv')
df5 = pd.read_csv('../Data/Group_data/weekdays_weekend_jan.csv')

```

```python
weekdays_sum=df['Weekdays'].sum() * 2/5
weekend_sum=df['Weekend'].sum()

weekdays_sum1=df1['Weekdays'].sum() * 2/5
weekend_sum1=df1['Weekend'].sum()

weekdays_sum2=df2['Weekdays'].sum() * 2/5
weekend_sum2=df2['Weekend'].sum()

weekdays_sum3=df3['Weekdays'].sum() * 2/5
weekend_sum3=df3['Weekend'].sum()

weekdays_sum4=df4['Weekdays'].sum() * 2/5
weekend_sum4=df4['Weekend'].sum()

weekdays_sum5=df5['Weekdays'].sum() * 2/5
weekend_sum5=df5['Weekend'].sum()

```

```python
# make a line chart for weekdays_sum and weekend_sum
plt.plot([weekdays_sum,weekdays_sum1,weekdays_sum2,weekdays_sum3,weekdays_sum4,weekdays_sum5],label='Weekdays')
plt.plot([weekend_sum,weekend_sum1,weekend_sum2,weekend_sum3,weekend_sum4,weekend_sum5],label='Weekend')
plt.legend()
plt.xlabel('Months')

plt.xticks([0,1,2,3,4,5],['Aug','Sep','Oct','Nov','Dec','Jan'])
plt.ylabel('Number of Trips')
plt.title('Number of bevarages sold')
plt.show()
```
