```python
# weekdays vs weekend days different data pointer
df1=df

df1['Date'] = pd.to_datetime(df['Date'],format='%d-%m-%Y')
df1['Weekday'] = df['Date'].dt.day_name()
df1.head()

# group by date and sum the total
df1 = df.groupby(['Date','Weekday']).sum()
df1.head()

#plot the graph
df1.plot(kind='bar',figsize=(20,10))
plt.show()
```

```python
#group by weekday and sum the total
df1 = df.groupby(['Weekday']).sum()
df1.head()
```

```python
# group by item and sum the total
df2=df.groupby(['Item name']).sum()
df2 
```

```python
# sort by quantity
df2.sort_values(by=['Quantity'],ascending=False,inplace=True)
df2
```

```python
# monday all item quantity
df3 = df.loc[df['Weekday'] == 'Monday']
df3.sort_values(by=['Quantity'],ascending=False,inplace=True)
df3
```

```python
# count item name 
df3=df3['Item name'].value_counts()
df3
```

```python
# sum of df3 to df7
df8 = df3.add(df4, fill_value=0)
df8 = df8.add(df5, fill_value=0)
df8 = df8.add(df6, fill_value=0)
df8 = df8.add(df7, fill_value=0)
df8
df8.to_csv('../Data/weekdays_Aug.csv',encoding="utf-8")
```
